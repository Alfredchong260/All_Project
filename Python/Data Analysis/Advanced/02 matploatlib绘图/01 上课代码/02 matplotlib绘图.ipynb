### 多坐标轴图表

```python
x = [i for i in range(1, 61)]
y_shanghai = [random.uniform(15, 18) for i in x]
y_beijing = [random.uniform(13, 16) for i in x]

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20, 8), dpi=100)
"""
plt.subplots 指定nrows=1, ncols=2, 一行两列
    fig --> 指代整个图片对象
    axes --> 指代的是子图对象
    
    axes[0]  第一个子图对象
    axes[1]  第二个子图对象
"""

# plt.plot(x, y_shanghai, label='上海', color='green')
axes[0].plot(x, y_shanghai, label='上海', color='green')
axes[1].plot(x, y_beijing, label='北京', color='red')

# 显示图例
# plt.legend(loc=0)
axes[0].legend()
axes[1].legend()

x_ticktes_label = [f'11点{i}分' for i in x]
y_ticktes = [i for i in range(40)]  # 准备y轴的数据范围
# plt.xticks(x[::5], x_ticktes_label[::5])
# plt.yticks(y_ticktes[::5])
axes[0].set_xticks(x[::5])
axes[0].set_xticklabels(x_ticktes_label[::5])
axes[0].set_yticks(y_ticktes[::5])

axes[1].set_xticks(x[::5])
axes[1].set_xticklabels(x_ticktes_label[::5])
axes[1].set_yticks(y_ticktes[::5])

axes[0].grid()
axes[1].grid()

# plt.xlabel('时间/分', fontsize=20)
# plt.ylabel('温度/摄氏度', fontsize=20)
# plt.title('某城市11点到12点1小时内每分钟的温度变化折线图')
axes[0].set_xlabel('时间/分', fontsize=20)
axes[0].set_ylabel('温度/摄氏度', fontsize=20)
axes[0].set_title('某城市11点到12点1小时内每分钟的温度变化折线图')

axes[1].set_xlabel('时间/分', fontsize=20)
axes[1].set_ylabel('温度/摄氏度', fontsize=20)
axes[1].set_title('某城市11点到12点1小时内每分钟的温度变化折线图')

plt.savefig('多坐标轴绘制.png')
```

## 分组柱状图

```python
movies = {
    "流浪地球": [2.01, 4.59, 7.99, 11.83, 16],
    "飞驰人生": [3.19, 5.08, 6.73, 8.10, 9.35],
    "疯狂的外星人": [4.07, 6.92, 9.30, 11.29, 13.03],
    "新喜剧之王": [2.72, 3.79, 4.45, 4.83, 5.11],
    "廉政风云": [0.56, 0.74, 0.83, 0.88, 0.92],
    "神探蒲松龄": [0.66, 0.95, 1.10, 1.17, 1.23],
    "小猪佩奇过大年": [0.58, 0.81, 0.94, 1.01, 1.07],
    "熊出没·原始时代": [1.13, 1.96, 2.73, 3.42, 4.05]
}
```

```python

"""构建每一天数据"""
movie_data = list(movies.values())

every_day = []
for i in range(len(movie_data[0])):
    every_day.append([
        movie_data[0][i],
        movie_data[1][i],
        movie_data[2][i],
        movie_data[3][i],
        movie_data[4][i],
        movie_data[5][i],
        movie_data[6][i],
        movie_data[7][i],
    ])

"""图表绘制"""
plt.figure(figsize=(20, 8), dpi=100)
# 指定柱子的宽度
width = 0.75  # 每一个大柱子的宽度
bin_width = width/5  # 每一个小柱子的宽度

# 生成电影数据的索引
index = range(len(movies))  # 01234567

# # 绘制第一天的数据
# plt.bar([i - bin_width * 2 for i in index], every_day[0], width=bin_width, label='第一天')

# day_tiket = every_day[0]
# # 往柱子上面添加数据需要循环一个一个加
# xs = [i - bin_width * 2 for i in index]

# for x , ticket in zip(xs, day_tiket):             # x - 0.08 往左移动一点
#                                                   # ticket + 0.05  往上移动一点
#     plt.annotate(ticket, xy=(x , ticket), xytext=(x - 0.08, ticket + 0.05))

# 第二种方案
for ind in range(len(every_day)):
    # 遍历每一天的数据
    day_tickets = every_day[ind]
    # 构建 x 位置的索引
    xs = [i - (bin_width * (2 - ind)) for i in index]
    # 绘制柱状图
    plt.bar(xs, day_tickets, width=bin_width, label=f"第{ind + 1}天")

    # 添加坐标上的数字
    # 需要一个个绘制数据
    for ticket, x in zip(day_tickets, xs):
        # ticket 2.01 每一天的数据
        # 1 - 2 * bin_width
        plt.annotate(ticket, xy=(x, ticket), xytext=(x - 0.08, ticket + 0.1))
    
plt.xticks(index, movies.keys())
plt.legend()
plt.grid()
```
