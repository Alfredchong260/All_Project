# 散点图

```python
import matplotlib.pyplot as plt
from random import random

plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

## 基本散点图

```python
plt.figure(figsize=(20, 8))

x = [random()  for i in range(10)]
y = [random()  for j in range(10)]


plt.scatter(x, y)
```

## 设置散点大小

```python
size = [(20 * random()) ** 2 for i in range(10)]

color = ['red', '#ffd700', '#1f77b4', 'red', '#ffd700', '#1f77b4', 'green', 'red', '#ffd700', '#1f77b4']

# s 设置点的大小, c设置点的颜色
plt.scatter(x, y, s=size, c=color, alpha=1, marker='P')

```

## 绘制多组数据

```python
x = [random()  for i in range(10)]
y = [random()  for j in range(10)]

x1 = [random()  for i in range(10)]
y1 = [random()  for j in range(10)]


# s 设置点的大小, c设置点的颜色
plt.scatter(x, y,marker='P', label='第一组')
plt.scatter(x1, y1, marker='X', label='第二组')
plt.legend()
```

# 饼图

## 绘制饼图

```python
edu = [59104, 1467937, 3579974]
labels = ['第一产业', '第二产业', '第三产业']

explod = [0, 0.2, 0]  # 

plt.pie(
    x=edu,
    labels=labels,
    autopct='%.3f%%',  # 显示数据的百分比
    labeldistance=0.8,  # label距离图像的距离
    colors=['red', 'green', 'pink'],  # 指定颜色
    explode=explod,  # 突出显示饼状块
    textprops={'fontsize': 20, 'color': 'blue'}
    
)
```

## 案例-学历分布饼图

```python
plt.figure(figsize=(10,8))
# 构造数据
edu = [0.2515,0.3724,0.3336,0.0368,0.0057]
labels = ['中专','大专','本科','硕士','其他']

explode = [0,0.1,0,0,0]  # 用于突出大专
colors=['#FEB748','#EDD25D','#FE4F54','#51B4FF','#dd5555'] # 自定义颜色


plt.pie(
    x=edu,
    autopct='%.2f%%',
    labels=labels,
    colors=colors,
    explode=explode
    
)
```

# 雷达图

```python
import numpy as np


properties = ['输出', 'KDA', '发育', '团战', '生存', '输出']
values = [40, 91, 44, 90, 95, 40]

plt.figure(figsize=(6,6))

theta = np.linspace(0, np.pi * 2, 6)  # 准备 0 - 2π 等分成六分的等分数据

plt.polar(theta, values)
plt.xticks(theta, properties)
plt.fill(theta, values)
```
